## Handling Missing Values with Mode-Based Imputation

The code below performs missing value imputation on two key columns in the dataset: `'Rating'` and `'Number_of_Ratings'`. The chosen fill values (5.0 and 1.0, respectively) indicate that these are likely the **mode** of their respective columns. This suggests the data is **skewed**, making mean or median imputation less appropriate.

---

````python
#-----------------------------------
# Filling missing values in the 'Rating' column with 5.0
# The choice of 5.0 suggests that this is the mode or most frequent rating
# This approach is typically used when the distribution is skewed and imputing with the mean or median is not appropriate
#-----------------------------------
df['Rating'] = df['Rating'].fillna(5.0)

#-----------------------------------
# Filling missing values in the 'Number_of_Ratings' column with 1.0
# Again, the use of 1.0 suggests it's the most frequent value in the distribution
# This method assumes that the column is heavily skewed, likely with many entries having a low count of ratings
#-----------------------------------
df['Number_of_Ratings'] = df['Number_of_Ratings'].fillna(1.0)
````

## Encoding the 'Product_Name' Column Using Label Encoding

This code applies **label encoding** to the `'Product_Name'` column, which converts categorical string values into numerical labels. This is particularly useful when the column has a large number of unique values, making one-hot encoding impractical.

---

````python
#-----------------------------------
# Initializing a LabelEncoder instance to convert categorical text labels into numeric labels
# This approach is used here because 'Product_Name' has many unique values
# Using label encoding avoids the curse of dimensionality that would occur if one-hot encoding was applied
#-----------------------------------
label_enc = LabelEncoder()

#-----------------------------------
# Transforming the 'Product_Name' column using label encoding
# The original string labels are replaced with integer labels ranging from 0 to n-1
#-----------------------------------
df['Product_Name'] = label_enc.fit_transform(df['Product_Name'])
````

## One-Hot Encoding the 'Category' Column

This code applies **one-hot encoding** to the `'Category'` column, transforming a categorical feature with a small number of unique values into multiple binary indicator columns. One-hot encoding is ideal here because the column contains only five unique categories.

````python
#-----------------------------------
# Initializing OneHotEncoder to convert the 'Category' column into one-hot encoded features
# sparse_output=False ensures the output is a dense NumPy array instead of a sparse matrix
# One-hot encoding is suitable here because 'Category' has only 5 unique values
#-----------------------------------
one_hot_enc = OneHotEncoder(sparse_output=False)

#-----------------------------------
# Fitting the encoder on the 'Category' column and transforming it into one-hot encoded format
# The result is a 2D NumPy array where each column corresponds to one category
#-----------------------------------
category_encoded = one_hot_enc.fit_transform(df[['Category']])

#-----------------------------------
# Creating new column names for each one-hot encoded feature, prefixed by 'Category_'
#-----------------------------------
category_cols = [f"Category_{cat}" for cat in one_hot_enc.categories_[0]]

#-----------------------------------
# Creating a new DataFrame from the one-hot encoded array with appropriate column names and original DataFrame index
#-----------------------------------
df_one_hot_enc = pd.DataFrame(category_encoded, columns=category_cols, index=df.index)

#-----------------------------------
# Concatenating the new one-hot encoded columns back to the original DataFrame
#-----------------------------------
df = pd.concat([df, df_one_hot_enc], axis=1)

#-----------------------------------
# Dropping the original 'Category' column as it has now been encoded into multiple binary columns
#-----------------------------------
df.drop('Category', axis=1, inplace=True)
````

## Feature Engineering: Creating the 'Rating_Factor' Column

This code generates a new feature, `'Rating_Factor'`, which is the product of two existing columns: `'Rating'` and `'Number_of_Ratings'`. This feature aims to capture both the quality of the rating and the quantity of ratings a product has received.

---

````python
#-----------------------------------
# Creating a new feature called 'Rating_Factor' by multiplying 'Rating' and 'Number_of_Ratings'
# This feature combines the quality (rating) and quantity (number of ratings) into a single metric
# This can help capture the overall impact or reliability of a product's rating
#-----------------------------------
df['Rating_Factor'] = df['Rating'] * df['Number_of_Ratings']
````

## Anomaly Detection on Training Data Using One-Class SVM and Local Outlier Factor

This code snippet performs anomaly detection on the training data using two popular unsupervised methods: **One-Class SVM (OCSVM)** and **Local Outlier Factor (LOF)**. The goal is to identify and remove outliers before model training to improve model robustness and performance.

---

````python
#-----------------------------------
# Indicating the start of anomaly detection process on the training data
#-----------------------------------
print("Starting anomaly detection on training data...")

#-----------------------------------
# Initializing One-Class SVM with specified parameters
# - nu=0.05: An upper bound on the fraction of training errors (outliers)
# - kernel='rbf': Using the radial basis function kernel for flexibility in decision boundaries
# - gamma='scale': Kernel coefficient calculated based on feature variance
# Fitting the model and predicting inliers (+1) and outliers (-1)
#-----------------------------------
ocsvm = OneClassSVM(nu=0.05, kernel='rbf', gamma='scale')
ocsvm_preds = ocsvm.fit_predict(X_train)

#-----------------------------------
# Initializing Local Outlier Factor (LOF)
# - n_neighbors=20: Number of neighbors used for local density estimation
# - contamination=0.05: Expected proportion of outliers in the data
# LOF’s fit_predict method simultaneously fits and predicts inliers/outliers
#-----------------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
lof_preds = lof.fit_predict(X_train)

#-----------------------------------
# Creating boolean masks for inliers (True) and outliers (False) from model predictions
# Both models label inliers as +1 and outliers as -1
#-----------------------------------
ocsvm_inliers = (ocsvm_preds == 1)
lof_inliers = (lof_preds == 1)

#-----------------------------------
# Combining the masks to keep only data points labeled as inliers by BOTH models
# This conservative approach reduces false positives in anomaly detection
#-----------------------------------
combined_inliers = ocsvm_inliers & lof_inliers

#-----------------------------------
# Printing original and filtered training set sizes to observe the effect of anomaly removal
#-----------------------------------
print(f"Original training set size: {len(X_train)}")
print(f"Training set after removing anomalies: {combined_inliers.sum()}")

#-----------------------------------
# Filtering the training features and target variables to keep only inliers
# This results in a “clean” training dataset free of detected anomalies
#-----------------------------------
X_train_clean = X_train[combined_inliers]
y_train_clean = y_train[combined_inliers]
````

## Hyperparameter Tuning of Random Forest Regressor Using GridSearchCV

This code performs a grid search to find the optimal hyperparameters for a Random Forest regression model, using 5-fold cross-validation and RMSE as the evaluation metric.

---

````python
#-----------------------------------
# Defining the hyperparameter grid for Random Forest Regressor
# Parameters to tune:
# - n_estimators: number of trees in the forest (100, 200, 300)
# - max_depth: maximum depth of each tree (None means unlimited depth)
# - min_samples_split: minimum samples required to split an internal node (2, 5, 10)
# - min_samples_leaf: minimum samples required to be at a leaf node (1, 2, 4)
#-----------------------------------
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

#-----------------------------------
# Instantiating the Random Forest Regressor with a fixed random state for reproducibility
#-----------------------------------
rf = RandomForestRegressor(random_state=42)

#-----------------------------------
# Setting up GridSearchCV to find the best hyperparameters
# - estimator: the model to tune (Random Forest Regressor)
# - param_grid: dictionary of parameters and their candidate values
# - scoring: negative root mean squared error, as GridSearchCV expects a score (higher is better)
# - cv: number of cross-validation folds (5-fold CV)
# - n_jobs: use all available CPU cores for parallel processing
# - verbose: level of logging output (2 for detailed)
#-----------------------------------
grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid_rf,
    scoring='neg_root_mean_squared_error',
    cv=5,
    n_jobs=-1,
    verbose=1
)

#-----------------------------------
# Printing a message to indicate the start of the grid search process
#-----------------------------------
print("Starting GridSearchCV for Random Forest...")

#-----------------------------------
# Fitting the GridSearchCV object on the cleaned training data
# This performs an exhaustive search over specified hyperparameter values
#-----------------------------------
grid_rf.fit(X_train_clean, y_train_clean)

#-----------------------------------
# Displaying the best hyperparameters found by GridSearchCV
#-----------------------------------
print("Best Random Forest parameters found:")
print(grid_rf.best_params_)

#-----------------------------------
# Displaying the best cross-validated RMSE score (convert from negative)
#-----------------------------------
print(f"Best CV RMSE: {-grid_rf.best_score_:.4f}")
````

## Hyperparameter Tuning for XGBoost Regressor Using GridSearchCV

---

````python
#-----------------------------------
# Defining the hyperparameter grid for XGBoost Regressor
# Parameters to tune:
# - n_estimators: number of boosting rounds (100, 200, 300)
# - max_depth: maximum depth of each tree (3, 6, 9)
# - learning_rate: step size shrinkage to prevent overfitting (0.01, 0.05, 0.1)
# - subsample: fraction of training instances used per tree (0.8, 1.0)
# - colsample_bytree: fraction of features used per tree (0.8, 1.0)
#-----------------------------------
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

#-----------------------------------
# Initializing the XGBoost Regressor with a fixed random state and squared error objective
#-----------------------------------
xgb = XGBRegressor(random_state=42, objective='reg:squarederror')

#-----------------------------------
# Configuring GridSearchCV for hyperparameter tuning on XGBoost
# - scoring: negative root mean squared error for evaluation
# - cv: 5-fold cross-validation
# - n_jobs: parallelize across all CPU cores
# - verbose: detailed logging during fit process
#-----------------------------------
grid_xgb = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid_xgb,
    scoring='neg_root_mean_squared_error',
    cv=5,
    n_jobs=-1,
    verbose=1
)

#-----------------------------------
# Indicating the start of the grid search for XGBoost
#-----------------------------------
print("Starting GridSearchCV for XGBoost...")

#-----------------------------------
# Fitting the grid search object to the cleaned training data
#-----------------------------------
grid_xgb.fit(X_train_clean, y_train_clean)

#-----------------------------------
# Printing the best hyperparameters found during the search
#-----------------------------------
print("Best XGBoost parameters found:")
print(grid_xgb.best_params_)

#-----------------------------------
# Printing the best cross-validation RMSE score (converted from negative)
#-----------------------------------
print(f"Best CV RMSE: {-grid_xgb.best_score_:.4f}")
````

## Functions for Preprocessing New Input Data and Predicting/Appending Results

This section defines two functions:

---

### 1. `preprocess_input(new_record: dict) -> pd.DataFrame`

- Converts a new input record (provided as a Python dictionary) into a pandas DataFrame.
- Applies the **same preprocessing steps** that were used on the training data:
  - Label encodes the `Product_Name` column using the previously fitted LabelEncoder.
  - One-hot encodes the `Category` column using the previously fitted OneHotEncoder.
  - Drops the original categorical columns after encoding.
- Ensures that the resulting DataFrame has **all the columns** in the exact order as the training dataset.
- Fills any missing columns (not present in the new record) with zeros to maintain feature consistency.

---

### 2. `predict_and_append(new_record: dict, model, dataset_path: str) -> dict`

- Takes a new input record and a trained model as input.
- Preprocesses the input using `preprocess_input`.
- Uses the model to predict the target variable (`Price` in this context).
- Rounds and formats the prediction as a currency string.
- Appends the new record along with its prediction to an existing CSV dataset file at the specified path.
  - Creates the CSV if it doesn't already exist.
- Returns a dictionary containing the formatted prediction string.

---

### Usage Context

- These functions facilitate **streamlined inference** for new data points.
- The `predict_and_append` function helps maintain an updated dataset by logging all new predictions.
- Ensures data consistency by applying identical preprocessing pipelines as used during training.

---

````python
def preprocess_input(new_record: dict) -> pd.DataFrame:
    #-----------------------------------
    # Convert input dictionary to DataFrame with a single row
    #-----------------------------------
    df_new = pd.DataFrame([new_record])

    #-----------------------------------
    # Scaling data
    #-----------------------------------
    df_new[['Rating', 'Number_of_Ratings']] = scaler.transform(df_new[['Rating', 'Number_of_Ratings']])

    
    #-----------------------------------
    # Label encode the 'Product_Name' column using the fitted LabelEncoder
    #-----------------------------------
    df_new['Product_Name_Encoded'] = label_enc.transform(df_new['Product_Name'])

    #-----------------------------------
    # One-hot encode the 'Category' column using the fitted OneHotEncoder
    #-----------------------------------
    cat_encoded = one_hot_enc.transform(df_new[['Category']])
    cat_cols = [f"Category_{cat}" for cat in one_hot_enc.categories_[0]]
    df_cat_one_hot_enc = pd.DataFrame(cat_encoded, columns=cat_cols, index=df_new.index)

    #-----------------------------------
    # Concatenate the one-hot encoded category columns and drop original 'Category' and 'Product_Name'
    #-----------------------------------
    df_processed = pd.concat([df_new.drop(['Category', 'Product_Name'], axis=1), df_cat_one_hot_enc], axis=1)

    #-----------------------------------
    # Add any missing columns that exist in training data but not in this new data,
    # filling them with zeros to maintain consistency in feature space
    #-----------------------------------
    missing_cols = set(X_train.columns) - set(df_processed.columns)
    for col in missing_cols:
        df_processed[col] = 0

    #-----------------------------------
    # Reorder columns to exactly match training data feature order
    #-----------------------------------
    df_processed = df_processed[X_train.columns]

    return df_processed


def predict_and_append(new_record: dict, model, dataset_path: str) -> dict:
    #-----------------------------------
    # Apply preprocessing to input dictionary
    #-----------------------------------
    df_input = preprocess_input(new_record)

    #-----------------------------------
    # Predict price using the provided model, extract first (and only) prediction
    # Round the predicted value and format it as a price string
    #-----------------------------------
    prediction_raw = model.predict(df_input)[0]
    prediction_rounded = round(prediction_raw, 2)
    prediction_formatted = f"${prediction_rounded}"

    #-----------------------------------
    # Prepare record with original input data and append predicted price
    #-----------------------------------
    record_to_append = new_record.copy()
    record_to_append['Predicted_Price'] = prediction_formatted

    #-----------------------------------
    # Check if dataset CSV file exists:
    # - If yes, read existing data and append new record then save
    # - If no, create new CSV with this record
    #-----------------------------------
    if os.path.exists(dataset_path):
        df_existing = pd.read_csv(dataset_path)
        df_existing = pd.concat([df_existing, pd.DataFrame([record_to_append])], ignore_index=True)
        df_existing.to_csv(dataset_path, index=False)
    else:
        pd.DataFrame([record_to_append]).to_csv(dataset_path, index=False)

    #-----------------------------------
    # Return dictionary containing the formatted prediction
    #-----------------------------------
    return {'prediction': prediction_formatted}
````
